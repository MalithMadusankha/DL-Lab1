```python
import matplotlib.pyplot as plt

# Generate random 100000x1 arrays from exponential, uniform, and normal distributions
scale = 1.0
exp_data = np.random.exponential(scale, size=(100000, 1))
exp_data
```

```python
uniform_data = np.random.uniform(low=0.0, high=1.0, size=(100000, 1))
uniform_data
```

```python
normal_data = np.random.normal(loc=0.0, scale=1.0, size=(100000, 1))
normal_data
```

```python

# Plot the histograms
plt.figure(figsize=(14, 7))

# Histogram for Exponential distribution
plt.subplot(1, 3, 1)
plt.hist(exp_data, bins=100, color='blue', alpha=0.7)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram for Uniform distribution
plt.subplot(1, 3, 2)
plt.hist(uniform_data, bins=100, color='green', alpha=0.7)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram for Normal distribution
plt.subplot(1, 3, 3)
plt.hist(normal_data, bins=100, color='red', alpha=0.7)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Adjust layout for better viewing
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid for x and y
X, Y = np.meshgrid(x, y)

# Define the function Z = X^2 + Y^2
Z = X**2 + Y**2

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D plot of Z = X^2 + Y^2')

# Show the plot
plt.show()
```

```python
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
import pandas as pd
```

```python
link = './Pokemon.csv'
# Read dataset
df = pd.read_csv(link, index_col=0, encoding='latin')
df.head(5)
```

```python
# Inspect the columns to find the correct names
print(df.columns)
```

```python
# Select the features of interest
features = ['HP', 'Attack', 'Defense', 'Sp_atk', 'Sp_def', 'Speed']
stats_df = df[features]
stats_df.sample(2)
```

```python
# Calculate correlations
corr = stats_df.corr()
corr
```

```python
# Heatmap
plt.figure(figsize=(9,8))
sns.heatmap(corr)
```

```python
# Scatterplot arguments
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=True, # No regression line
           hue='Generation')   # Color by evolution stage
```
