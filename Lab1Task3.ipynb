```python
import pandas as pd
import matplotlib.pyplot as plt

dataset_link = './Au_nanoparticle_dataset.csv'

df = pd.read_csv(dataset_link)
df.sample(2)
print("Hello world")
```

```python
features = ['N_total', 'N_bulk', 'N_surface']
filters = df[feature]
filters
```

```python
filters.head(20)
```

```python
# Calculate mean, standard deviation and quartile values
stats = filters.describe().loc[['mean', 'std', '25%', '50%', '75%']]
```

```python

# Set up the matplotlib figure with 1x4 layout
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Plot the histogram for each feature
for i, feature in enumerate(features):
    axes[i].hist(features[feature], bins=20, color='blue', alpha=0.7)
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```
